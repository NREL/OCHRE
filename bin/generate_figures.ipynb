```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(6, 5))

# Simulated tradeoff curve
x = np.linspace(0.01, 1, 100)
y = 1 / x

# Cap the curve at the maximum setpoint value
y = np.clip(y, 0, 10)

# Fill under the curve to indicate energy use
ax.fill_between(x, y, color='lightblue', alpha=0.6, label='Hot Water Availability')

# Plot the curve outline
ax.plot(x, y, color='black', label='Hot Water Delivered')

# Add arrows to axes
ax.annotate('', xy=(1.05, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))
ax.annotate('', xy=(0, 11), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))

# Add chart title
ax.text(0.5, 11.5, 'Energy/ Hot Water Tradeoff', fontsize=14, ha='center', weight='bold')

# Add axis labels
ax.text(-0.1, 5, 'Energy Use', fontsize=12, rotation=90, va='center')
ax.text(0.5, -1, 'Unmet Demand', fontsize=12, ha='center')

# Annotate the maximum setpoint line
ax.axhline(y=10, color='gray', linestyle='--', linewidth=1)
ax.text(0.6, 10.3, 'Maximum Setpoint', fontsize=10, color='gray')

# Add legend
ax.legend(loc='upper right', bbox_to_anchor=(1, 0.8), fontsize=10, frameon=False)


# Set axis limits and remove ticks and spines
ax.set_xlim(-0.05, 1.1)
ax.set_ylim(-1, 12)
ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(6, 5))

# Simulated tradeoff curve
x = np.linspace(0.01, 1, 100)
y = 1 / x

# Cap the curve at the maximum setpoint value
y = np.clip(y, 0, 10)

# Fill under the curve to indicate energy use
ax.fill_between(x, y, color='lightblue', alpha=0.6, label='Energy Use')

# Plot the curve outline
ax.plot(x, y, color='black')

# Add arrows to axes
ax.annotate('', xy=(1.05, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))
ax.annotate('', xy=(0, 11), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))

# Add chart title
ax.text(0.5, 11.5, 'Energy/ Hot Water Tradeoff', fontsize=14, ha='center', weight='bold')

# Add axis labels
ax.text(-0.1, 5, 'Energy Use', fontsize=12, rotation=90, va='center')
ax.text(0.5, -1, 'Cold Water Draws', fontsize=12, ha='center')

# Annotate the maximum setpoint line
ax.axhline(y=10, color='gray', linestyle='--', linewidth=1)
ax.text(0.6, 10.3, 'Maximum Setpoint', fontsize=10, color='gray')

# Annotate unmet demand
ax.text(0.7, 1.5, 'Unmet Demand', fontsize=10, color='black')

# Set axis limits and remove ticks and spines
ax.set_xlim(-0.05, 1.1)
ax.set_ylim(-1, 12)
ax.axis('off')

plt.tight_layout()
plt.show()
```
