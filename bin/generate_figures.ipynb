```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 20})

# Create figure and axis
fig, ax = plt.subplots(figsize=(6, 5))

# Simulated tradeoff curve
x = np.linspace(0.01, 1, 100)
y = x * 10
y_rand = np.random.rand(100) * 5

# Cap the curve at the maximum setpoint value
y = np.clip(y, 0, 10)

# Fill *above* the curve to the top (e.g., 12), representing unmet demand
ax.fill_between(x, y, 10, color='lightblue', alpha=0.6, label='Unmet Demand')

# Plot the curve outline
ax.plot(x, y, color='black', label='Hot Water Delivered')

# Add arrows to axes
ax.annotate('', xy=(1.05, 0), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))
ax.annotate('', xy=(0, 11), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', linewidth=1.5, color='black'))

# Title and axis labels
ax.text(0.5, 11.5, 'Energy/ Hot Water Tradeoff', fontsize=16, ha='center', weight='bold')
ax.text(-0.1, 5, 'Energy Use', fontsize=16, rotation=90, va='center')
ax.text(0.5, -1, 'Hot Water Availability', fontsize=16, ha='center')

# Setpoint lines
ax.axhline(y=10, color='gray', linestyle='--', linewidth=1)
ax.text(0.6, 10.3, 'Maximum Setpoint', fontsize=16, color='gray')

ax.axhline(y=5, color='gray', linestyle='--', linewidth=1)
ax.text(0.6, 5.3, 'Ideal Setpoint', fontsize=16, color='gray')

# Vertical line at intersection with Ideal Setpoint (y=5)
ax.axvline(x=0.5, ymin=0.1, ymax=0.8, color='red', linestyle='--', linewidth=1)
ax.text(0.8, 0.3, 'Hot Water Consumed', fontsize=13, color='red', ha='center')

# Legend
ax.legend(loc='upper right', bbox_to_anchor=(1.2, 0.4), fontsize=13, frameon=False)
#ax.legend(loc='upper left', bbox_to_anchor=(1.05, 0.75))

# Axis limits and cleanup
ax.set_xlim(-0.05, 1.1)
ax.set_ylim(-1, 12)
ax.axis('off')

plt.tight_layout()
plt.show()

```
